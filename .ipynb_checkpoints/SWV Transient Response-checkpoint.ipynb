```python
# Second-order transient response model e/ exponential term
# lig_C is a global variable used to pass probe concentration values to the model, as that isn'tt a variable being solved for
def first_time_response(x, A1, A2, k_1, k_2, k_3, k_4, lig_I, b, c):
    
     # composite coefficient 
    K_A = (lig_C*k_1)+k_2
    K_B = (lig_I*k_3)+k_4
    
    K_F = 0.5*(K_A+K_B+math.sqrt(((K_A-K_B)**2)+(4*lig_C*lig_I*k_1*k_3)))
    K_S = 0.5*(K_A+K_B-math.sqrt(((K_A-K_B)**2)+(4*lig_C*lig_I*k_1*k_3)))
    
    # composite terms within the equation
    # this is a very long equation that is easy to make mistakes on
    steady_state = (A1*lig_C*k_1)/(K_F-K_S)
    term1 = (k_4*(K_F-K_S))/(K_F*K_S)
    term2 = ((k_4-K_F)/K_F)*np.exp(-1*K_F*x)
    term3 = ((k_4-K_S)/K_S)*np.exp(-1*K_S*x)
    term4 = A2*x*np.exp(-1*x/c)
    
    # The actual equation
    y = steady_state*(term1+term2-term3)-term4+b
    return y

def f_first_time_response(x, p):
    return first_time_response(x, *p)
```

```python
# Second-order transient response model
# lig_C is a global variable used to pass probe concentration values to the model, as that isn'tt a variable being solved for
# lig_I is concentration of third non-specific variable - not a global variable - model is optimizing for this variable
def second_time_response(x, A, k_1, k_2, k_3, k_4, lig_I, b):
    
    # composite coefficient 
    K_A = (lig_C*k_1)+k_2
    K_B = (lig_I*k_3)+k_4
    
    K_F = 0.5*(K_A+K_B+math.sqrt(((K_A-K_B)**2)+(4*lig_C*lig_I*k_1*k_3)))
    K_S = 0.5*(K_A+K_B-math.sqrt(((K_A-K_B)**2)+(4*lig_C*lig_I*k_1*k_3)))
    
    # composite terms within the equation
    # this is a very long equation that is easy to make mistakes on
    steady_state = (A*lig_C*k_1)/(K_F-K_S)
    term1 = (k_4*(K_F-K_S))/(K_F*K_S)
    term2 = ((k_4-K_F)/K_F)*np.exp(-1*K_F*x)
    term3 = ((k_4-K_S)/K_S)*np.exp(-1*K_S*x)
    
    # The actual equation
    y = steady_state*(term1+term2-term3)+b
    return y

def f_second_time_response(x, p):
    return second_time_response(x, *p)
```

```python
# Extracts target concentration value from file name
# returns target concentration in M
def get_lig_C(data_name):
    
    # Extracting the value and SI multiplier from the third word in the title
    target_conc = data_name.split() 
    value = float(target_conc[2][:-2])
    SI_symbol = target_conc[2][-2]
    
    possible_SI_symbol = ["m", "u", "n", "p"] # micro needs to be 'u'
    possible_SI_mult = [1000, 1, 0.001, 0.000001]
    
    # since function returns value in uM, value is adjusted based on SI symbol in title
    for i in range(len(possible_SI_symbol)):
        if possible_SI_symbol[i] == SI_symbol:
            return value*possible_SI_mult[i]
        else:
            continue
            
    return
```

```python
def fit_curvefit(p0, datax, datay, function, yerr, **kwargs):
    """
    Note: As per the current documentation (Scipy V1.1.0), sigma (yerr) must be:
        None or M-length sequence or MxM array, optional
    Therefore, replace:
        err_stdev = 0.2
    With:
        err_stdev = [0.2 for item in xdata]
    Or similar, to create an M-length sequence for this example.
    """
    pfit, pcov = \
         curve_fit(function, datax, datay, p0, bounds=(0, np.inf), \
                            sigma=yerr, maxfev = 500000000, **kwargs)
    error = [] 
    for i in range(len(pfit)):
        try:
          error.append(np.absolute(pcov[i][i])**0.5)
        except:
          error.append(0.00)
    pfit_curvefit = pfit
    perr_curvefit = np.array(error)
    return pfit_curvefit, perr_curvefit 
```

```python
# Initial predictions for the two models
# 1st order model parameters
one_A1 = 5 #2.47852108 #2.47828412
one_A2 = 0.25
one_k_1 = 0.21348682 #0.51350722
one_k_2 =  0.89481087 #0.89478663
one_k_3 =  5.68174284 #5.68729447
one_k_4 = 0.02671008 #0.0267102
one_lig_I = 0.01971486 #0.0196961
one_b = 0.45017237 #0.045017234
one_c = 6

# 2nd order model parameters
two_A = 2.47852108  #2.47828412
two_k_1 = 0.21348682 #0.51350722
two_k_2 =  0.89481087 #0.89478663
two_k_3 =  5.68174284 #5.68729447
two_k_4 = 0.02671008 #0.0267102
two_lig_I = 0.01971486 #0.0196961
two_b = 0.45017237 #0.045017234

i = 2

# target concnentration extracted from file name
lig_C = get_lig_C(data_names_list[i])

y1_pred = first_time_response(data_list[i]["Time"].values, one_A1, one_A2, one_k_1, one_k_2, one_k_3, one_k_4, one_lig_I, one_b, one_c)
y2_pred = second_time_response(data_list[i]["Time"].values, two_A, two_k_1, two_k_2, two_k_3, two_k_4, two_lig_I, two_b)
data_list[i].insert(2, "y1_pred", y1_pred, True)
data_list[i].insert(3, "y2_pred", y2_pred, True)

# fitting the data

err_stdev = [0.1 for item in data_list[i]["Time"].values]

one_popt, one_perr = fit_curvefit(
    function = first_time_response,
    datax = data_list[i]["Time"].values,
    datay = data_list[i]["Peak Current"].values,
    p0 = [one_A1, one_A2, one_k_1, one_k_2, one_k_3, one_k_4, one_lig_I, one_b, one_c],
    yerr=err_stdev
)
two_popt, two_perr = fit_curvefit(
    function = second_time_response, 
    datax = data_list[i]["Time"].values, 
    datay = data_list[i]["Peak Current"].values,
    p0 = [two_A, two_k_1, two_k_2, two_k_3, two_k_4, two_lig_I, two_b],
    yerr=err_stdev
)

# one_popt, one_pcov = curve_fit(
#     f = first_time_response,
#     xdata = data_list[i]["Time"].values,
#     ydata = data_list[i]["Peak Current"].values,
#     p0 = [one_A1, one_A2, one_k_1, one_k_2, one_k_3, one_k_4, one_lig_I, one_b, one_c],
#     bounds = (0, np.inf),
#     maxfev = 50000000,
# )
# two_popt, two_pcov = curve_fit(
#     f = second_time_response, 
#     xdata = data_list[i]["Time"].values, 
#     ydata = data_list[i]["Peak Current"].values,
#     p0 = [two_A, two_k_1, two_k_2, two_k_3, two_k_4, two_lig_I, two_b],
#     bounds = (0, np.inf),
#     maxfev = 500000000,
# )

# Calculating y-values with kinetics function based on fitted model parameters
y1_model = first_time_response(data_list[i]["Time"].values, *one_popt)
y2_model = second_time_response(data_list[i]["Time"].values, *two_popt)
data_list[i].insert(4, "y1_model", y1_model, True)
data_list[i].insert(5, "y2_model", y2_model, True)

# function for Calculating variance of the fit (R^2)
# takes model fit y-values as input
# measured y-values are global
def get_r_squared(y_model):
    residuals = data_list[i]['Peak Current'] - y_model
    ss_res = np.sum(residuals**2)     # residual sum of square
    ss_tot = np.sum((data_list[i]['Peak Current']- np.mean(data_list[i]['Peak Current']))**2)     # total sum of squares
    r_squared = 1 - (ss_res/ss_tot)
    return r_squared

# Find r_squared and inserting the value into the dataframes
one_r_squared = get_r_squared(data_list[i]['y1_model'])
two_r_squared = get_r_squared(data_list[i]['y2_model'])
data_list[i]['1_R-squared'] = pd.Series([one_r_squared])
data_list[i]['2_R-squared'] = pd.Series([two_r_squared])

# Inserting model parameters into the dataframe
data_list[i]["y1_legend"] = pd.Series(["A1", "A2", "k_1", "k_2", "k_3", "k_4", "lig_I", "b", "c"])
data_list[i]["y1_param"] = pd.Series(one_popt.tolist())
data_list[i]["y2_legend"] = pd.Series(["A", "k_1", "k_2", "k_3", "k_4", "lig_I", "b"])
data_list[i]["y2_param"] = pd.Series(two_popt.tolist())

# Std deviation errors on the fitted parameters
one_perr = np.sqrt(np.diag(one_pcov))
two_perr = np.sqrt(np.diag(two_pcov))

# plotting original measurements as scatter plot with model fittings
# green is first order fitting
# red is second order fitting
fig, ax = plt.subplots(figsize=(6, 6))
p1 = sns.scatterplot(data=data_list[i], x="Time", y="Peak Current", ax=ax)
p2 = sns.lineplot(data=data_list[i], x="Time", y="y1_pred", color="green", linestyle='dashed', ax=ax)
p3 = sns.lineplot(data=data_list[i], x="Time", y="y1_model", color="green", ax=ax)
p4 = sns.lineplot(data=data_list[i], x="Time", y="y2_pred", color="red", linestyle='dashed', ax=ax)
p5 = sns.lineplot(data=data_list[i], x="Time", y="y2_model", color="red", ax=ax)
fig.suptitle(data_names_list[i], fontsize=18)
ax.set_xlabel("Time (min)")
ax.set_ylabel("Peak Current (A)")
if generate_output:
    plt.savefig(cwd.joinpath("Successful Fits", "#"+str(i)+"_"+data_names_list[i]+".png"), dpi=300)

# Report for curve fitting
print(data_names_list[i])
print("\nModel 1 (1st-order) Fitted Parameters:")
print(two_popt)
print("\nModel 1 (1st-order) Fitting Errors:")
print(one_perr)
print("\nModel 1 (1st-order) R^2:")
print(one_r_squared)
print("\nModel 2 (2nd order) Fitted Parameters:")
print(one_popt)
print("\nModel 2 (2nd-order) Fitting Errors:")
print(two_perr)
print("\nModel 2 (2st-order) R^2:")
print(two_r_squared)

# saves fitted values to csv alongside the original values 
# save as csv file separated by tabs
if generate_output:
    data_list[i].to_csv(cwd.joinpath("Successful Fits", "#"+str(i)+"_"+data_names_list[i]+"_MODEL.csv"),
                        sep=',', encoding='utf-8')
```
